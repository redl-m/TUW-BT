# Dataset Augmentation


```python
import pandas as pd
import numpy as np

# Load existing dataset
DATA_PATH = "../data/training/training_data.csv"
df = pd.read_csv(DATA_PATH)

# Soft skills to insert
soft_skills = [
    'Structural Adherence',
    'Adaptive Fluidity',
    'Interpersonal Influence',
    'Execution Velocity',
    'Psychological Resilience'
]

# Inject random scores (1 to 5) for the soft skills
np.random.seed(42) # Ensures reproducibility
for skill in soft_skills:
    # Use a normal distribution centered at 3 to mimic realistic human populations
    raw_scores = np.random.normal(loc=3.0, scale=1.0, size=len(df))
    df[skill] = np.clip(np.round(raw_scores), 1, 5).astype(int)

# Inject 'Job Hopping' metric
job_hopping_categories = ['Low', 'Medium', 'High']
df['Job Hopping'] = np.random.choice(job_hopping_categories, size=len(df), p=[0.6, 0.3, 0.1]) # 'Low' is most common, 'High' is rare

# Link the new features to the AI Score so the Random Forest learns them
resilience_bonus = (df['Psychological Resilience'] - 3) * 2
hopping_penalty = np.where(df['Job Hopping'] == 'High', -6,
                           np.where(df['Job Hopping'] == 'Medium', -2, 0))

# Apply modifiers and ensure the final score doesn't break the 0-100 boundary
df['AI Score (0-100)'] = np.clip(df['AI Score (0-100)'] + resilience_bonus + hopping_penalty, 0, 100)

# Save augmented dataset
AUGMENTED_PATH = "../data/training/training_data.csv"
df.to_csv(AUGMENTED_PATH, index=False)
print(f"Successfully injected features and saved to: {AUGMENTED_PATH}")
```
